```python
import cv2
from PIL import Image
#from google.colab.patches import cv2_imshow
from pedestrian_detection_ssdlite import Ped_BBox_Module as ped_Mod
from matplotlib import pyplot as plt
```

```python
def detect_video(video_path):
  #cap = cv2.VideoCapture(0)
  cap = cv2.VideoCapture(video_path)
  
  while cap.isOpened():
    ret, frame = cap.read()
    #print(ret,frame)
    if ret:
        bbox_list = ped_Mod.get_person_bbox(frame, thr=0.5)
        #print(bbox_list)
        for i in bbox_list:
          cv2.rectangle(frame, i[0], i[1], (125, 255, 51), thickness=3)
          cv2.putText(frame, text="Pedestrian", org=i[0], fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                    fontScale=0.50, color=(36,255,12), thickness=2)
              
        #h, w, c = frame.shape
        #print(frame.shape)
                
        cv2.imshow("detected", cv2.resize(frame,(650,420)))
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
        
    else:
        break
  cap.release()
  cv2.destroyAllWindows()          
```

```python
ped_vd = input("Enter Pedestrian Video = ")
detect_video(ped_vd)
```
